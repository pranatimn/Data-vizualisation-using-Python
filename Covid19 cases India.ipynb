# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

# Get data

```python
df = pd.read_csv("raw_data1.csv")
```

```python
df.head()
```

```python
data = df.iloc[:,:]
data
```

# Inspect the dataframe

```python
data.info
```

# Inspect null values in each column

```python
round(data.isnull().sum().sort_values(ascending=False)/len(data)*100,2)
```

# Inspect null values in each row

```python
data.isnull().sum(axis=1).sort_values(ascending= False)
```

# Total No of Covid-19 Corona Virus Patient in India Gender wise

```python
data.groupby('Gender')['Num Cases'].sum()
```

# Which age group is affected the most

```python
M = data.groupby('Age Bracket')['Num Cases'].sum().sort_values(ascending= False).head(10)
M
```

```python
M.plot.bar()
plt.show()
```

# Total No of Covid-19 Corona Virus Patient in India State wise

```python
M = data.groupby('Detected State')['Num Cases'].sum().sort_values(ascending=False)
M
```

```python
M.plot.bar(figsize=(15,5))
plt.show()
```
